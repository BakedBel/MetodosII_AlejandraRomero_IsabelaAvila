```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def System(N,t): ,
            NA= N[0]
    NB= N[1]
    NC= N[2]
    dNAdt = (-0.5)*NA
    dNBdt = (0.5)*NA - (0.3)*NB
    dNCdt = (0.3)*NB

    return np.array([dNAdt,dNBdt,dNCdt])
```

```python
N0=[500,0,0]

t=np.linspace(0,10,50)

solucion=odeint(System,N0,t)

plt.scatter(t,solucion[:,0],label='NA')
plt.scatter(t,solucion[:,1],label='NB')
plt.scatter(t,solucion[:,2],label='NC')
plt.legend()
```

```python
def Increment(f,N0,t,h):
    NA=N0[0]
    NB=N0[1]
    NC=N0[2]

    R=np.array(([NA,NB,NC]))
    K1=f(R,t)

    R=np.array(([NA+0.5*h*K1[0],NB+0.5*h*K1[1],NC+0.5*h*K1[2]]))
    K2=f(R,t+0.5*h)

    R=np.array(([NA+0.5*h*K2[0],NB+0.5*h*K2[1],NC+0.5*h*K2[2]]))
    K3=f(R,t+0.5*h)

    R=np.array(([NA+h*K3[0],NB+ h*K3[1],NC+ h*K3[2]]))
    K4=f(R,t+h)

    Delta=(h*(K1 + 2*K2 + 2*K3 + K4))/6

    return N0+Delta

def Adaptativo(f,N0,t,e=1e-4):
    h = t[1] - t[0]

    tf = t[-1]
    t= t[0]
    r=N0
    Vectors=r
    TimeVector = np.array([t])


    while t < tf:

        r1 = Increment(f,r,t,h)
        r2 = Increment(f,r1,t+h,h)
        rt = Increment(f,r,t,2*h)


        dx,h= checkpasos(r1,r2,rt,h)

        t += h

        r = Increment(f,r,t,h)

        TimeVector = np.append(TimeVector,t)
        Vectors = np.vstack((Vectors,r))

    return TimeVector, Vectors



def checkpasos(r1,r2,rt,h,e=1e-4):


    error = np.linalg.norm( r2 - rt )

    if error < e:
        h=2*h
        dx = rt


    else:
       q = 0.85*(e/error)**(1/4)
       h = h*q
       dx=0

    return dx,h

```
