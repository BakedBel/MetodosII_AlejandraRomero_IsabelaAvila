```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_exacta(t, alfa):
    resultado = np.e**(t * alfa)
    return resultado

def funcion_aproximada(delta, k, alfa, u0):
    resultado = ((1 + delta)**k) * u0
    return resultado


tiempo_exacto = np.arange(0, 10, 0.01)
valores_k = np.arange(0, 6, 1)
alfa = -1
u0 = funcion_exacta(0, alfa)

plt.plot(tiempo_exacto, funcion_exacta(tiempo_exacto, alfa), label='Función exacta')
plt.plot(valores_k, funcion_aproximada(1.1, valores_k, alfa, u0), label='dt=1.1')
plt.plot(valores_k, funcion_aproximada(1.5, valores_k, alfa, u0), label='dt=1.5')
plt.plot(valores_k, funcion_aproximada(1.9, valores_k, alfa, u0), label='dt=1.9')
plt.legend()

plt.xlabel('Tiempo / k')
plt.ylabel('Función')
plt.show()
```
