```python
import numpy as np
import matplotlib.pyplot as plt

#Calcular aceleraci칩n
def a(x):
    w = np.pi
    return (-x) * (w ** 2)

def metodo_verlet(t, x, x0, v0):
    #Definir paso
    h = t[1] - t[0]
    #Primer paso, valores del problema
    x[0] = x0
    #Segundo valor, se calcula con Euler
    x[1] = x[0] + v0 * h
    
    for i in range(2, len(t)):
        aceleracion = a(x[i-1])
        x[i] = 2 * x[i-1] - x[i-2] + aceleracion * (h ** 2)

t = np.linspace(0, 10, 100)
x = np.zeros(len(t))
x0 = 1
v0 = 1

tin = np.arange(0,10,3/np.pi)
xin = np.zeros(len(tin))
metodo_verlet(t, x, x0, v0)
metodo_verlet(tin,xin,x0,v0)

# Crear el gr치fico de doble eje
fig = plt.figure(figsize=(10,5))
ax1= fig.add_subplot(121)
ax2= fig.add_subplot(122)

# Plotear la temperatura
ax1.plot(t,x, 'r-')
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Posici칩n')
ax1.set_title('Estable')

# Crear el segundo eje y
ax2.plot(tin,xin, 'b-')
ax2.set_xlabel('Tiempo (s)')
ax2.set_ylabel('Posici칩n')
ax2.set_title('Inestable')


```
