```python
import numpy as np
import matplotlib.pyplot as plt

#Calcular aceleración
def a(x):
    w = np.pi
    return (-x) * (w ** 2)

#función exacta
def f(t):
    return (np.cos(np.pi*t) + (np.sin(np.pi*t) )/ np.pi)

def metodo_verlet(t, x, x0, v0):
    #Definir paso
    h = t[1] - t[0]
    #Primer paso, valores del problema
    x[0] = x0
    #Segundo valor, se calcula con Euler
    x[1] = x[0] + v0 * h
    
    for i in range(2, len(t)):
        aceleracion = a(x[i-1])
        x[i] = 2 * x[i-1] - x[i-2] + aceleracion * (h ** 2)

t = np.linspace(0, 10, 100)
x = np.zeros(len(t))
x0 = 1
v0 = 1

metodo_verlet(t, x, x0, v0)
x_teorica = f(t)

plt.plot(t, x, label="Método de Verlet", linestyle='--', color='b')
plt.plot(t, x_teorica, label="Solución Teórica", linestyle='-', color='r')

plt.xlabel("Tiempo (s)")
plt.ylabel("Distancia (m)")
plt.legend()
plt.grid(True)
plt.show()
```
