```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Function(x):
    return (1/x**2) -(2/(x**2 + 2))

def System(y,x):
    return x*y**2 - (2*y/x) - (1/x**3)

def Integrator(f,y0,x):
    h=x[1]-x[0]
    y=np.zeros_like(x)
    y[0]=y0
    k1=np.zeros(1)
    k2=np.zeros(1)

    R=np.zeros(1)

    for i in range(1,len(x)):
        R=np.array([y[i-1]])

        k1=f(R,x[i-1])

        R=np.array([y[i-1]+h*k1[0]])

        k2=f(R,x[i-1]+h)

        y[i]= y[i-1] + 0.5*h*(k1[0] + k2[0])

    return y
```

```python
x=np.linspace(1.41,15,100)
y0=0
s= Function(x)

y=Integrator(System,y0,x)

plt.plot(x,s,label='Exacta')
plt.plot(x,y,label='Runge Kutta 2')

plt.legend()
```

Podemos observar que la aproximación de Runge kutta es muy buena, ya que no se puede ver la diferencia entre la función exacta y su aproximación.
